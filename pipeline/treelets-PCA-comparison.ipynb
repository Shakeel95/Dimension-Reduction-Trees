## Setup

```r
library("treelet") # treelet implimentation
library("corrplot") # graphical display of correlation matrices
```

## Data

```r
url = "https://raw.githubusercontent.com/Shakeel95/ST499-Tilting-/master/clean%20UK%20macro%20data.csv"
UK_macro_data = read.csv(url, stringsAsFactors=FALSE)
```

## Real Data

```r
C = cov(as.matrix(UK_macro[1:83,]))
cc = cor(as.matrix(UK_macro[1:83,]))
```

```r
corrplot(cc)
```

```r
tree = Run_JTree(X = C, 
                 maxlev = 49, # maximum height of tree
                 whichsave = c(1,2,3,20,49) # save basis functions for these levels 
                )
```

```r
C = cov2cor(tree$TreeCovs[[49]])
corrplot(corr = C, method = "color")
```
